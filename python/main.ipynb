```python
import mitsuba as mi
mi.set_variant('cuda_ad_rgb')
```

```python
scene = mi.load_file('../assets/Scenes/dragon/scene_onesided.xml')
img = mi.render(scene)
mi.util.write_bitmap('out.png', img)
mi.Bitmap(img)
```

```python
# direct lighting
scene = mi.load_file('../assets/Scenes/cbox.xml')

img = mi.render(scene, spp=256)
mi.util.write_bitmap('out.png', img, quality=1  )
mi.Bitmap(img)
```
